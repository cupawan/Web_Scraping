```python
import requests
from bs4 import BeautifulSoup

# bhaskar_url = "https://www.bhaskar.com/career/"
# bhaskar_url = "https://www.bhaskar.com/ayodhya-ram-mandir/"
# bhaskar_url = "https://www.bhaskar.com/rashifal/"
# bhaskar_url = "https://www.bhaskar.com/"
bhaskar_url = "https://www.bhaskar.com/db-original/"

response = requests.get(bhaskar_url)

soup = BeautifulSoup(response.content, 'html.parser')

headlines = soup.find_all(class_='c7ff6507')

all_urls_dict = {}
all_urls = []

for i in headlines:
    for j in i.contents:
        all_urls_dict[j.text] = []
        slug = j.get('href')
        if slug:
            all_urls_dict[j.text].append(bhaskar_url+slug)
            all_urls.append(bhaskar_url+slug)
            
for key,value in all_urls_dict.items():
    headline = key
    news = ''''''
    for i in value:
        response = requests.get(i)
        soup = BeautifulSoup(response.content, 'html.parser')
        news_element = soup.find_all('p')
        for n in news_element:
            news += n.text
    print("### " + headline + " ###")
    print(news)
    print("-"*50)
    
```
